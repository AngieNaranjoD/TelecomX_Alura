### Otra forma de normalizar los datos podría ser:

```python
# Crear DataFrame inicial con las columnas simples
df_base = datos_telecomx[['customerID', 'Churn']].copy()
```

```python
# Normalizar datos anidados de 'customer'
customer_df = pd.json_normalize(datos_telecomx['customer'])
customer_df.columns = ['customer_' + col for col in customer_df.columns]
```

```python
# Normalizar datos anidados de 'phone'
phone_df = pd.json_normalize(datos_telecomx['phone'])
phone_df.columns = ['phone_' + col for col in phone_df.columns]
```

```python
# Normalizar datos anidados de 'internet'
internet_df = pd.json_normalize(datos_telecomx['internet'])
internet_df.columns = ['internet_' + col for col in internet_df.columns]
```

```python
# Normalizar datos anidados de 'account' (incluyendo charges anidados)
account_df = pd.json_normalize(datos_telecomx['account'])
# Renombrar columnas para mayor claridad
account_df.columns = [
    col.replace('Charges.', 'account_charges_') if 'Charges.' in col
    else 'account_' + col
    for col in account_df.columns
]
```

```python
# Combinar todos los DataFrames
df_final = pd.concat([df_base, customer_df, phone_df, internet_df, account_df], axis=1)
df_final.sample()
```
